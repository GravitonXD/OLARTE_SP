# PREPROCESSOR - DATA EXPLORATION

In this notebook, we will explore the collected data to better understand how we might able to use it to train our deep learning model.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## PSEI

```python
# Load PSEI data
df_psei = pd.read_csv('data/PSEI.csv')
df_psei.head(10)
```

```python
# We also need to check for any missing values
df_psei.isnull().sum()
```

Since there is no missing data in the PSEI dataset, we will start by exploring the data.

```python
# Describe the data
df_psei.describe()
```

```python
# Starting Date from Data
min_date = df_psei['Date'].min()
print(f"Minumum Date: {min_date}")
```

```python
# Ending Date from Data
max_date = df_psei['Date'].max()
print(f"Maximum Date: {max_date}")
```

```python
# Plot the Data (Open, High, Low, Close)
plt.figure(figsize=(16,8))
plt.title('PSEI (Opening Prices))')
plt.xlabel('Days')
plt.ylabel('Opening Prices')
plt.plot(df_psei['Open'])
plt.show()

plt.figure(figsize=(16,8))
plt.title('PSEI (High Prices))')
plt.xlabel('Days')
plt.ylabel('High Prices')
plt.plot(df_psei['High'])
plt.show()

plt.figure(figsize=(16,8))
plt.title('PSEI (Low Prices))')
plt.xlabel('Days')
plt.ylabel('Low Prices')
plt.plot(df_psei['Low'])
plt.show()

plt.figure(figsize=(16,8))
plt.title('PSEI (Closing Prices))')
plt.xlabel('Days')
plt.ylabel('Closing Prices')
plt.plot(df_psei['Close'])
plt.show()
```

```python
# Putting all the data in one plot
plt.figure(figsize=(16,8))
plt.title('PSEI (Opening, High, Low, Closing Prices))')
plt.xlabel('Days')
plt.ylabel('Prices')
plt.plot(df_psei['Open'])
plt.plot(df_psei['High'])
plt.plot(df_psei['Low'])
plt.plot(df_psei['Close'])
plt.legend(['Open', 'High', 'Low', 'Close'], loc='upper left')
plt.show()
```
