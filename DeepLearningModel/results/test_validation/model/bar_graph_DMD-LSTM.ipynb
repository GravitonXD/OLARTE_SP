```python
import matplotlib.pyplot as plt
```

```python
#MSE, RMSE, MAE, MAPE
win5 = [31669.602700854284,177.95955355320007,128.60952210660952,0.01911102535927252]
win10 = [75216.39203344575,274.25607018522993,189.90278989583334,0.02842209857651895]
win15 = [120814.02153661667,347.58311457350266,239.8749610595703,0.03606343363790969]
win20 = [162395.82288064936,402.9836508850568,280.75467301359305,0.04235908956264941]
sizes = [5,10,15,20]
```

```python
# Bar Graph for MSE across different window sizes
plt.bar([5,10,15,20], [win5[0],win10[0],win15[0],win20[0]], color='blue')
plt.title('MSE across different window sizes (DMD-LSTM)')
plt.xlabel('Window Size')
plt.ylabel('MSE')
plt.xticks([5,10,15,20])
# Add a bar on top showing the difference between the two (next bar - current bar)
plt.bar([5,10,15,20], [win10[0]-win5[0],win15[0]-win10[0],win20[0]-win15[0],0], bottom=[win5[0],win10[0],win15[0],win20[0]], color='red')
plt.ylim(0, 175000)
plt.show()

for i in range(4):
    diff = "{:.5f}".format([win10[0]-win5[0],win15[0]-win10[0],win20[0]-win15[0],0][i])
    try:
        print(f'Difference between window size {sizes[i]} and {sizes[i+1]} is {diff}')
    except:
        pass
```
