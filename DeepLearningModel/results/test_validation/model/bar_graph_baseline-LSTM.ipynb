```python
import matplotlib.pyplot as plt
```

```python
#MSE, RMSE, MAE, MAPE
win5 = [33503.81428096558,183.04047170220463,129.00945773990597,0.019163145083920433]
win10 = [75141.48322538845,274.11946889155547,189.72945121527778,0.028412556281991217]
win15 = [124298.02785032894,352.5592543819109,242.56766749790737,0.03644370305864025]
win20 = [162011.71304742625,402.50678633710794,280.05673017096416,0.04230858353880549]
sizes = [5, 10, 15, 20]
```

```python
# Bar Graph for MSE across different window sizes
plt.bar([5,10,15,20], [win5[0],win10[0],win15[0],win20[0]], color='blue')
plt.title('MSE across different window sizes (Baseline LSTM)')
plt.xlabel('Window Size')
plt.ylabel('MSE')
plt.xticks([5,10,15,20])
# Add a bar on top showing the difference between the two (next bar - current bar)
plt.bar([5,10,15,20], [win10[0]-win5[0],win15[0]-win10[0],win20[0]-win15[0],0], bottom=[win5[0],win10[0],win15[0],win20[0]], color='red')
plt.ylim(0, 175000)
plt.show()

for i in range(4):
    diff = "{:.5f}".format([win10[0]-win5[0],win15[0]-win10[0],win20[0]-win15[0],0][i])
    try:
        print(f'Difference between window size {sizes[i]} and {sizes[i+1]} is {diff}')
    except:
        pass
```
