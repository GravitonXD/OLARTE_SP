```python
import pandas as pd
```

```python
df = pd.read_csv('model_s5.csv', index_col=0)
df = df.sort_index()
df
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np
```

```python
# Plot the MSE distribution and add a normal distribution on top
plt.hist(df['MSE'], bins=5, density=True, alpha=0.6, color='g')
mu, std = norm.fit(df['MSE'])
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)
title = "MSE distribution. mu = %.2f,  std = %.2f" % (mu, std)
plt.title(title)
plt.show()
```
