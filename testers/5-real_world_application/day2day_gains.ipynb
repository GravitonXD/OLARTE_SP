```python
# A graph comparing the day-to-day gains between alamSYS and PSEI
import matplotlib.pyplot as plt
```

```python
alamsys = [-0.36, -0.25, 2.99, -0.03, -0.23, -0.49, 0, -0.15, -0.25, 0.28]
psei = [-1.290, -1.060, -0.770, -0.980, -3.34, -0.72, -2.060, -0.930, -1.32, -1.34]
```

```python
# Line Graph
plt.figure(figsize=(10, 6))
plt.plot(alamsys, label='AlamSYS')
plt.plot(psei, label='PSEI')
# Plot the difference in red dashed line
plt.plot([a - b for a, b in zip(alamsys, psei)], 'r--', label='Difference')
# Label each point in the difference line
for i in range(0, 10):
    plt.text(i, alamsys[i] - psei[i], str(round(alamsys[i] - psei[i], 2)))
plt.xlabel('Trading Days')
plt.xticks(range(0, 10), range(1, 11))
plt.ylabel('Gains (%)')
plt.yticks(range(-3,4))
plt.legend()
plt.show()
```
