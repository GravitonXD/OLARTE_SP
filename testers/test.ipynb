```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./test_logs/stress_test1.csv")
```

```python
df
```

```python
# transform total_time to epoch time
df["total_time"] = df["total_time"].apply(lambda x: pd.to_datetime(x).timestamp())
```

```python
# Plot the total_time vs run_number
plt.plot(df["run_number"], df["total_time"])
plt.xlabel("Run Number")
plt.ylabel("Total Time (s)")
plt.title("Total Time vs Run Number")
# Plot a line for the mean
plt.axhline(df["total_time"].mean(), color="red", linestyle="--")
# Plot a line for the median
plt.axhline(df["total_time"].median(), color="green", linestyle="--")
# Plot a line for the max
plt.axhline(df["total_time"].max(), color="orange", linestyle="--")
# Plot a line for the min
plt.axhline(df["total_time"].min(), color="blue", linestyle="--")
# Legend
plt.legend(["Total Time", "Mean", "Median", "Max", "Min"])
plt.show()
```
