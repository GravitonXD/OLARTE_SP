```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Open cpu_stats and memory_stats files
cpu_stats = pd.read_csv('./processed_data/cpu_stats.csv')
memory_stats = pd.read_csv('./processed_data/memory_stats.csv')
```

```python
# Plot the CPU stats
plt.figure(figsize=(20,10))
plt.plot(cpu_stats['alamAPI_API'], label='alamAPI_API')
plt.plot(cpu_stats['alamAPI_DB'], label='alamAPI_DB')
plt.legend()
plt.xlabel('Time (seconds)')
plt.ylabel('CPU Usage (%)')
plt.show()
```

```python
# Plot the Memory stats
plt.figure(figsize=(20,10))
plt.plot(memory_stats['alamAPI_API'], label='alamAPI_API')
plt.plot(memory_stats['alamAPI_DB'], label='alamAPI_DB')
plt.legend()
plt.xlabel('Time (seconds)')
plt.ylabel('Memory Usage (MiB)')
plt.show()
```
