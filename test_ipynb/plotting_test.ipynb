```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Plot the close price of the stock
df = pd.read_csv("./BDO.csv")

# Plot the close price of the stock
# show a trend line
df['Close'].plot(label='BDO(close)', figsize=(16,8))
df['Open'].plot(label='BDO(open)', figsize=(16,8))
(df['Close']-df['Open']).plot(label='BDO(close-open)', figsize=(16,8))
plt.legend()
plt.show()

plt.figure(figsize=(16,8))
plt.title('Close Price History')
plt.plot(df['Close']-df['Open'])
plt.xlabel('Date', fontsize=18)
plt.ylabel('Close Price PHP', fontsize=18)
plt.show()
```
