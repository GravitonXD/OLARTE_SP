Using Yahoo Finance Python Module, we will request for a copy of the historical data of the all the available stocks in the PH Market

```python
import yfinance as yf
```

```python
ticker = yf.Ticker("PSEI.PS")
ticker.info
```

```python
# Get the historical prices for this ticker
hist = ticker.history(period="max")
hist
```

```python
# Put the data into a csv file
hist.to_csv('PSEI.csv')
```

Now we know that everything is working correctly we can loop through all possible stocks in the Philippine Market and save their corresponding price historical data to ./historical_data

```python
# Create a list of all the tickers in the PSE from Yahoo Finance
PSE_tickers = [
    'PSEI.PS',
    'BDOUY',
    'AYALY',
    'JBFCF',
    'SMGBF',
]

for i in PSE_tickers:
    ticker = yf.Ticker(i)
    hist = ticker.history(period="max")
    hist.to_csv(f"{i}.csv")
```
