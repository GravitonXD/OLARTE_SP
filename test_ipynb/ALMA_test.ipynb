```python
# Import the data
df = pd.read_csv("./DITO.csv")

# Create a new data frame with the calculated ALMA values
df['ALMA'] = df['Close'].rolling(window=9).mean() + 2 * df['Close'].rolling(window=9).std()

# Delete rows that have NaN values
df = df.dropna()

# Print the data frame
print(df)

# Plot the data
plt.figure(figsize=(15, 5))
plt.plot(df['Date'], df['Close'], label='Close')
plt.plot(df['Date'], df['ALMA'], label='ALMA')
plt.legend(loc='upper left')
plt.show()
```
